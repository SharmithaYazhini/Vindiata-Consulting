**1.Which aircraft types are best suited for their operation?**


1.   Aircraft types, whose range is less than the distance between the cities should be ignored.
2.   Number of trips=Pass Demand / Number of seats
3.   Total cost= Cost per seat per km* Distance * Number of Seats* Number of trips



```python
for a,b,c,d in df1.itertuples(index=False):
  df7[str(a)+'-'+str(b)+' cost'] = np.where(df7["km"] >= d, (df7["Cost per seat per km"] * df7["Seats"] * d * np.ceil(c / df7["Seats"])).astype(int), "NaN")
  df7[str(a)+'-'+str(b)+' trip']=np.where(df7["km"] >= d,(np.ceil(c / df7["Seats"])).astype(int), "NaN")
```
